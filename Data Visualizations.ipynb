Now that we know what all of the data sets look like, lets start to do some data exploration and visualization. Distributions gives a general feel of some variables that are common to each data set. 

```python
plt.style.use('ggplot')
plt.hist(imdb_titlebasics['runtime_minutes'], edgecolor = 'black', bins = 50)
plt.title('Runtime Distribution (minutes)')
plt.xlabel('Run Time')

```

```python
# Create a figure to show distributions of variables examined.
fig = plt.figure()


plt.subplot(2, 2, 1)
plt.hist(imdb_titlebasics['runtime_minutes'], edgecolor = 'black', bins = 50, label = 'Runtime (minutes)')
plt.legend()

plt.subplot(2, 2, 2)
plt.hist(imdb_titleakas['averagerating'], edgecolor = 'black', bins = 50, label = 'Average Rating')
plt.legend()

plt.subplot(2, 2, 3)
plt.hist(tn['production_budget'], edgecolor = 'black', bins = 50, label = 'Production Budget')
plt.legend()

plt.subplot(2, 2, 4)
plt.hist(tn['domestic_gross'], edgecolor = 'black', bins = 50, label = 'Domestic Gross')
plt.legend()

plt.show()
```

```python
#use .corr() to try and visualize any pairwise correlations in the Rotten Tomatoe data set. 
rotten_tomatoes.corr()
```

```python
#lets get an idea of many movies are rated in each category. 
rotten_tomatoes['rating_x'].value_counts().plot(kind='bar')
plt.ylabel("Movie Rating Count")
plt.xlabel("Movie Rating")
```

```python
#rotten tomatoes uses a 'fresh' or 'rotten' review with 'fresh' aquired when a movie reaches %60 or more positive reviews. 
rotten_tomatoes['fresh'].value_counts().plot(kind='bar')
plt.ylabel("Rating Count")
plt.xlabel("Rating")
```

```python
rotten_tomatoes.corr()
```

```python
#lets see if movie ratings has any correlation to whethter it is considered 'fresh' or 'rotten'.
sns.set_style('whitegrid') 
sns.lmplot(x ='rating_num', y ='fresh_rotten',data = rotten_tomatoes,scatter_kws={'color': 'paleturquoise'})
```
